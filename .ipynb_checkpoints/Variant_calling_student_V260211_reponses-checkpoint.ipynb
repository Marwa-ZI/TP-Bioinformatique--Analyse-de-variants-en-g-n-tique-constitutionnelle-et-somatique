#### **Contexte clinique du Jupyter Notebook** : 
Un patient de 76 ans a été hospitalisé dans votre hôpital suite au diagnostic d'une leucémie aigue myéloide (prolifération de cellules tumorales, appelés blastes, dans la moelle osseuse). Nous avons à notre disposition un prélèvement de moelle osseuse fait lors du diagnostic (échantillon nommé "Diagnostic") et envahi par 30% de blastes. Le patient a également eu une biospie de peau avec culture des fibroblastes donc sans envahissement par des blastes (échantillon nommé "Germline").

#### **Objectif du Jupyter Notebook** : 
Après extraction de l'ADN pour chacun de ces échantillons, un séquençage ciblé sur un panel de 15 gènes a été fait en short-read en utilisant un sequenceur Illumina. Votre objectif est d'utiliser ces données de séquençage pour identifier les mutations somatiques acquises par les blastes ainsi que de potentielles mutations constitutionnelles chez ce patient.

### **3.** Appel des variants
#### Créez un nouveau dossier "Results"

```bash
# Code à écrire par l'étudiant
mkdir Results
```

#### On peut maintenant utiliser "VarScan" pour appeler les SNP (changement d'un nucléotide par un autre) et les indels (courtes insertion/délétion de nucléotides). Vous pouvez lancer la ligne de commande "varscan mpileup2cns --help" pour accéder à la documentation.

```bash
# Code déjà fourni
varscan mpileup2cns --help
```

#### Dans un premier temps nous allons utiliser VarScan avec un appel des variants volontairement très permissif.

```bash
# Code déjà fourni
varscan mpileup2cns \
Intermediate_files/Diagnosis.mpileup \
--min-coverage 1 \
--min-reads2 1 \
--min-var-freq 0.001 \
--min-avg-qual 1 \
--p-value 1 \
--strand-filter 0 \
--output-vcf 1 \
--variants \
> Results/Diagnosis.vcf
```

#### Combien de variants ont été appelés par VarScan ? Que pensez-vous de la pertinence de ces variants ?

#### Modifier la ligne de commande pour choisir des critères plus stringents afin de filtrer ces artefacts
Habituellement pour ces séquençages panels, le variant calling est réalisé sur les positions génomiques avec une profondeur d'au moins 10 reads. Les variants appelés doivent être soutenus par au moins 2 reads, représentant une fréquence allélique d'au moins 2%. Enfin, la qualité du variant doit être suffisamment bonne (généralement 20 est un bon threshold, et une p-value <1%) et on exclue les variants avec un biais de brin trop important.

```bash
## Code à écrire par l'étudiant
varscan mpileup2cns \
Intermediate_files/Diagnosis.mpileup \
--min-coverage 10 \
--min-reads2 2 \
--min-var-freq 0.02 \
--min-avg-qual 20 \
--p-value 0.01 \
--strand-filter 1 \
--output-vcf 1 \
--variants \
> Results/Diagnosis.vcf
```

#### Combien de variants ont été appelés par VarScan avec ces nouveaux critères ?

#### Affichez un extrait du fichier vcf avec la ligne de commande suivant et commentez le contenu

```bash
# Code déjà fourni
head -n 33 Results/Diagnosis.vcf | tail
```

#### Chargez le fichier vcf sur IGV pour et regardez manuellement les variants  identifiés

## **Sixième étape** : Annotation des variants en utilisant _VEP_

### L'outil VEP (Variant Effect Predictor) est développer par ensbl et permet d'annoter les variants identifiés.
#### Vous pouvez lancer la ligne de commande "vep --help" pour accéder à la documentation

```bash
# Code déjà fourni
/./opt/ensembl-vep/vep -help
```

```bash
# Code déjà fourni
/./opt/ensembl-vep/vep -i Results/Diagnosis.vcf \
    -o Results/Diagnosis_vep.vcf \
    --vcf \
    --database \
    --assembly GRCh38 \
    --fasta Reference_genome/hg38.fa \
    --pick \
    --symbol \
    --hgvs \
    --af \
    --polyphen b \
    --sift b
```

### A vous d'utiliser VEP pour annoter les variants identifiés sur l'échantillon de fibroblastes

```bash
# Code à écrire par l'étudiant
/./opt/ensembl-vep/vep -i Results/Germline.vcf \
    -o Results/Germline_vep.vcf \
    --vcf \
    --database \
    --assembly GRCh38 \
    --fasta Reference_genome/hg38.fa \
    --pick \
    --symbol \
    --hgvs \
    --af \
    --polyphen b \
    --sift b
```
